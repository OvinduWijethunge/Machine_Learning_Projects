```python
import pandas as pd
```

```python
df = pd.read_excel('new_commentData.xlsx',)
```

```python
pd.set_option('display.max_columns',None)
pd.set_option('display.max_rows',1000)
```

```python
df.head(2)
```

```python
li = df.values.tolist()
```

```python
du_list = []
di = dict()
count = 0
id_list = []
for i,val in enumerate(li):
    count +=1
    vid = val[0]
    cid=val[4]
    row_li = []
    mou = 0
    for ind,x in enumerate(li[count:]):
        if x[4]==cid and x[4]  not in id_list:
            mou +=1
            row_li.append(ind+count+2)
            row_li.append(x[0])
            row_li.append(x[4])
    
    if mou>0:
        
        st = "--{}----{}---{}---".format(i+2,vid,cid)
        
        du_list.append(st)        
        du_list.append(row_li)
        
            
        id_list.append(cid)   
    
```
